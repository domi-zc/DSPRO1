```python
from ultralytics import YOLO
```

```python
model = YOLO('yolo11s.pt')
```

```python
model.tune(data="../Dataset/data.yaml", imgsz=640, epochs=20)
```

```python
trained_model = YOLO("../../results/fine_tuned_best.pt")
```

```python
metrics = trained_model.val(
    data="../Dataset/data.yaml",
    split="test",
    save=True
)
```
